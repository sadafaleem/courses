**Data Selection using LOC and ILOC Methods**

*   LOC is label-based,meaning you have to specify based on column and row labels
*   ILOC is integer index based, you can specify rows and columns by their integer index



```python
# Example: Selecting Single Row
cars.loc["RU"]
```

```python
cars.iloc[4]
```

```python
# Example: Selecting Multiple Rows
cars.loc[["RU", "AUS"]]
```

```python
cars.iloc[[4,1]]
```
